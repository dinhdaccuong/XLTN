```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = 20
x_max = 5
F = 10
Fs = 100
n = np.arange(1, N+1)
```

```python
x_n = np.zeros(N)
print(x_n)
```

```python
i = 0
for i_n in n:
    x_n[i] = x_max*math.sin((2*math.pi*i_n*F)/Fs)
    i = i + 1
```

```python
def sign(x_n):
    if x_n > 0:
        return 1
    elif x_n == 0:
        return 0
    else:
        return -1
```

```python
V = max(x_n)
u = 255
y_n = np.zeros(N)
```

```python
i = 0
for xn in x_n:
    y_n[i] = V * (np.log(1+u*np.abs(xn)/V)/np.log(1+u))*sign(xn)
    i = i + 1
```

```python
plt.plot(x_n, label = "x(n)")
```

```python
plt.plot(y_n) 
```
