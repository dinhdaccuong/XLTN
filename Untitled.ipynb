```python
input_data = read("/home/cuong/XLTN/WAV/Xe.wav")
audio = input_data[1]
window = hann(1024)
audio = window * audio[0:1024]
#print(audio)14 mags = abs(rfft(audio))
# convert to dB
mags = abs(rfft(audio))
mags = 20 * scipy.log10(mags)
# normalise to 0 dB max
mags -= max(mags)
# plot
plt.plot(mags)
# label the axes
plt.ylabel("Magnitude (dB)")
plt.xlabel("Frequency Bin")
# set the title
plt.title("Flute Spectrum")
plt.show()
```
