```python
import scipy
import scipy.io.wavfile as wf
import numpy as np
import wave
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
from scipy.io.wavfile import read
import scipy.signal as sci
```

```python
rate, signal = read("/home/cuong/XLTN/WAV/Xe.wav")
print(rate, signal)
```

```python
signal_x = signal[7900:8200]
N = signal_x.shape[0]
K = signal_x.shape[0]
print(type(signal_x[9]))
signal_x = scipy.signal.resample(signal_x, 300)
print(type(signal_x[9]))
```

```python
n = 0
k = 0
Rk = np.zeros(N)
for k in range(0, K):
    for n in range(0, N-1-k):
         Rk[k] += signal_x[n] * signal_x[n+k]
```

```python
max_index = np.zeros(N)
max_value = np.zeros(N)
```

```python
max_index[0] = 0;
max_value[0] = Rk[0]
i = 0
j = 0
for i in range(0, K):
    if((i == K-1) and (Rk[i] > Rk[i-1])):
        max_index[j] = i;
        max_value[j] = Rk[i]
        j = j + 1
    elif((Rk[i] > Rk[i-1]) and (Rk[i] > Rk[i+1])):
        max_index[j] = i;
        max_value[j] = Rk[i]
        j = j + 1
```

```python
#finding max index of value
i = 1
max_i = 1
max_v = max_value[1]
for val in max_value[1:N]:
    if(val > max_v):
        max_i = i
        max_v = max_value[i]
    i = i + 1
```
